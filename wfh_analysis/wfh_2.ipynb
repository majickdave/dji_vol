```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('wfh_2.csv', parse_dates=['date'], skiprows=[0], usecols=range(1,11),
                 names=['bu', 'date', 'aht',
                        'handle_time', 'volume', 'conformance',
                       'adherence', 'abandonment rate', 'service level', 'unplanned out'])
df.head()
```

# Clean Data

```python
# reformat columns
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()

for kpi in ['conformance', 'adherence', 'abandonment rate']:
    df[kpi] = pd.to_numeric(df[kpi].str.replace('%', ''), errors='coerce')/100
    
df['aht'] = pd.to_numeric(df['aht'], errors='coerce', downcast='integer')
df['handle_time'] /= 3600
df['aht'] /= 60
```

```python
df.to_csv('clean_wfh_2.csv')
pd.read_csv('clean_wfh_2.csv', index_col=0)
```
