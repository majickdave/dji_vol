# Roll up forecast to the weekly, monthly, and quarterly level

```python
import os
import pandas as pd
from datetime import datetime


path = '../preds/'


for pred_file in ['aht']:
    preds_list = os.listdir(path+pred_file)
    for file_name in preds_list:
        df = pd.read_csv(path+pred_file+'/'+file_name, index_col=0, parse_dates=['ds'])
        future = df[df['ds'].dt.date > datetime.now().date()]
        print(file_name,pred_file+'\n')
#         display(future.groupby(df['ds'].dt.isocalendar().week).sum())
#         display(future.groupby(df['ds'].dt.month).sum())
#         display(future.groupby(df['ds'].dt.quarter).sum())
        display(future.groupby(df['ds'].dt.quarter).mean())
```
